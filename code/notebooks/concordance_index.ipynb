```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
from lifelines.utils import concordance_index
names = ['Alice', 'Bob', 'Carol', 'Dave', 'Eve']
events = [1, 2, 3, 4, 5]
preds = [1, 2, 3, 4, 5]
df = pd.DataFrame(data={'Churn times': events, 'Predictions': preds}, index=names)
print(df)
print(f'Concordance index: {concordance_index(events, preds)}')
```

```python
df = pd.read_csv('../../analyses/snipen_2021_RMS/2021_07_26_test_mock_data/7_compare_bracken_profiles/all_data.csv')
```

```python
events = df['true'].tolist()
preds = df['true'].tolist()
print(f'Concordance index: {concordance_index(events, preds)}')
```

```python
for col in df:
    if col.endswith('depth'):
        events = df[col].tolist()
        preds = df['true'].tolist()
        print(f'\n{col}')
        print(f'Concordance index: {concordance_index(events, preds)}')
        
        pearson_corr = df[col].corr(df['true'])
        print(f'Pearson correlation: {pearson_corr}')
```
