```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('./hhai_msei_100M_4_mah_distance.tsv', sep='\s+')
print(df.shape)
```

```python
# gen_df = pd.DataFrame()
# gen_df['genome_names'] = df.columns.to_list()
# gen_df.to_csv('genome_names.csv',index=None)

fragment_df = pd.read_csv('./../extract_read_features/helius_hhai_msei_100M.csv', index_col=0)
gen_ls = list(fragment_df['genome'].unique())
print(len(gen_ls))
gen_incl_ls = [i+'.fna' for i in gen_ls]
gen_excl_ls = [i for i in list(df.columns) if i not in gen_incl_ls]

# df = df.loc[gen_ls[0:113], gen_ls[0:113]]
df_incl = df.loc[gen_incl_ls, gen_incl_ls].copy()

df_excl = df.loc[gen_excl_ls, gen_excl_ls].copy()

print(df_incl.shape)
print(df_excl.shape)
```

```python
sns.set(font_scale=1)
# sns.clustermap(df_excl, figsize=(30,50), standard_scale=0, dendrogram_ratio=0.2, cmap="mako", metric='euclidean', yticklabels=True, xticklabels=False)
sns.clustermap(df_excl, figsize=(20,20), dendrogram_ratio=0.2, cmap="mako", annot=True, metric='euclidean', yticklabels=True, xticklabels=True)
plt.savefig('excl_distance.tif',dpi=350)
plt.show()
```

```python
sns.set(font_scale=0.5)
# sns.clustermap(df_excl, figsize=(30,50), standard_scale=0, dendrogram_ratio=0.2, cmap="mako", metric='euclidean', yticklabels=True, xticklabels=False)
sns.clustermap(df_incl, figsize=(5,5), standard_scale=0, dendrogram_ratio=0.2, cmap="mako", metric='euclidean', yticklabels=True, xticklabels=False)
plt.savefig('incl_distance.tif',dpi=350)
plt.show()
```
