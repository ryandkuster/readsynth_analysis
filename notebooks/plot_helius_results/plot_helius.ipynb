```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
taxa_total = 610
all_df = pd.read_csv('./helius_results.csv')
all_df['taxa_percent'] = (all_df['taxa_present'] / taxa_total) *100
all_df['percent_reads_lost'] = 100 - (all_df['read_final'] / all_df['read_sim'])*100
all_df['percent_reads_kept'] = (all_df['read_final'] / all_df['read_sim'])*100
all_df['percent_fragments_lost'] = 100 - (all_df['discernable_fragments'] / all_df['fragments_sim'])*100
all_df['percent_fragments_kept'] = (all_df['discernable_fragments'] / all_df['fragments_sim'])*100
all_df
```

```python

df = all_df[all_df['source'] == 'helius'].copy()
ten_M_df = all_df[all_df['read_target'] == 10000000].copy()
ten_M_df['sample'] = ten_M_df['source'] + "_" + ten_M_df['re1'] + "_" + ten_M_df['re2']
ten_M_df
```

```python
font = {'size'   : 14}

plt.rc('font', **font)
```

```python
fig, ax = plt.subplots()
ax.margins(0.05) # Optional, just adds 5% padding to the autoscaling
for group in list(df['sample'].unique()):
    ax.plot(df[df['sample']==group]['read_sim'], df[df['sample']==group]['taxa_percent'], marker='o', alpha=0.75, linestyle='', ms=10, label=group)
    ax.set_xscale('log')
ax.legend()
plt.xlabel('simulated reads (n)')
plt.ylabel('percent taxa identified')
fig.set_size_inches(10, 5)
plt.savefig('taxa_v_read_sim.tif', dpi=350)
plt.savefig('taxa_v_read_sim.png', dpi=350)
plt.show()
```
