## remove reads that map multiply from other genomes

```python
def get_hamming_distance(a, b):
    hamm = 0
    for i, j in zip(a, b):
        if i != j:
            hamm += 1
    return hamm/len(a)

threshold = 0.25 #rate of hamming distance
df_dupes = pd.DataFrame()

for i in range(0, df['length'].max()+1):
    tmp_df = df[df['length']==i].copy()
    seq_ls = tmp_df['seq'].to_list()
    gen_ls = tmp_df['genome'].to_list()
    
    hamm_ls = []
    for idxa, (seqa, gena) in enumerate(zip(seq_ls, gen_ls)):
        close_ones = 0
        for idxb, (seqb, genb) in enumerate(zip(seq_ls, gen_ls)):
            if idxa == idxb:
                pass
            elif gena == genb:
                pass
            else:
                hamm = get_hamming_distance(seqa, seqb)
                if hamm <= threshold:
                    close_ones += 1
        hamm_ls.append(close_ones)
    tmp_df.loc[:,'neighbors'] = hamm_ls
    df_dupes = pd.concat([df_dupes, tmp_df])
```

```python
df = df_dupes[df_dupes['neighbors'] == 0]
df
```

Calculate relative abundance by capturing the inter-taxa ratios within each fragment length. The average of these ratios will be used to determine the overall relative abundance of the taxa

```python
gen_ls=df['genome'].unique().tolist()

ratios_df = pd.DataFrame(0, index=gen_ls, columns=gen_ls)
copies_df = pd.DataFrame(0, index=gen_ls, columns=gen_ls)


def record_ratios(ratios_df, copies_df, counts_ls, gen_ls):
    for gena, counta in zip(gen_ls, counts_ls):
        for genb, countb in zip(gen_ls, counts_ls):
            if counta > 0 and countb > 0:
                ratios_df.loc[gena, genb] = (counta / countb) + ratios_df.loc[gena, genb]
                copies_df.loc[gena, genb] = copies_df.loc[gena, genb] + 1

    return ratios_df, copies_df


for i in range(1,df['length'].max()):
    print(i)
    counts_ls = []
    for j in gen_ls:
        depth = df[(df['length']==i)&(df['genome']==j)]['observed'].mean()
        if not math.isnan(depth):
            counts_ls.append(depth)
        else:
            counts_ls.append(0)
    ratios_df, copies_df = record_ratios(ratios_df, copies_df, counts_ls, gen_ls)
```

```python
for i in range(0,20):
    np_copies = np.array(copies_df)
    np_ratios = np.array(ratios_df)
    np_ratios = np_ratios/np_copies
    np_sums = np_ratios.sum(axis=0)
    # abundance_col = np_ratios[:, np.where(np_sums == np_sums.max())[0][0]]
    abundance_col = np_ratios[:, i]
#     abundance_col = np.mean(np.delete(np_ratios,16,1),axis=1)

    o_ls = []
    e_ls = []

    for i, j in zip(gen_ls, abundance_col):
        a = df[df['genome']==i]['rel_abund'].unique()[0]
        b = j/abundance_col.sum()
        a = '{0:.4f}'.format(a)
        b = '{0:.4f}'.format(b)
#         print(f'E: {a}  O: {b}')
        o_ls.append(float(b))
        e_ls.append(float(a))

    from scipy.stats.stats import pearsonr     
    print(pearsonr(e_ls,o_ls))
```
