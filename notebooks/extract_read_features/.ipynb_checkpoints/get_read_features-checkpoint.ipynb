```python
import numpy as np
import pandas as pd


def check_first_in_pair(flag):
    return bin(flag)[-7] == '1'


def check_in_cut_sites(site_dt, df_for):
    results = []
    np_for = np.array(df_for)

    for i in np_for:
        chr = i[2]
        try:
            results.append(i[13] in site_dt[chr])
        except KeyError:
            results.append(False)

    return results


final_file = './sort_fixmate_GCA_000005845.2_ASM584v2_genomic.sam'
with open(final_file) as f:
    for idx, line in enumerate(f):
        if line.startswith('@'):
            pass
        else:
            sam_header = idx
            break

df = pd.read_csv(final_file, skiprows=sam_header, usecols=range(11), sep='\s+', header=None)
cols = 'query flag ref start qual cigar pair pair_pos length seq score'.split()
df.columns = cols
df['end'] = df['start'] + df['seq'].str.len() - 1
df['forward'] = df['flag'].apply(lambda x: check_first_in_pair(x))
df = df.drop(['flag', 'qual', 'cigar', 'pair', 'pair_pos', 'score'], axis=1)
df_for = df[df['forward'] == True]
df_rev = df[df['forward'] == False]
df = df_for.merge(df_rev, on="query")
df['ecori_pos'] = np.where(df['length_x']>0, df['start_x']-2, df['end_x']-5)
df['msei_pos'] = np.where(df['length_x']<0, df['start_y']-2, df['end_y']-3)

cut_sites = pd.read_csv('./raw_digest_GCA_000005845.2_ASM584v2_genomic.fna.csv')
np_cut_sites = np.array(cut_sites)

count_ls = []

for i in np_cut_sites:
    c_start = i[2]
    c_end = i[3]
    c_m1 = i[4]
    c_m2 = i[5]
    if c_m1 == 'GAATTC':
        tmp_df_top = df[(df['ecori_pos']==c_start) & (df['msei_pos']==c_end)]
        count_ls.append(tmp_df_top.shape[0])
    elif c_m1 == 'TTAA':
        tmp_df_bot = df[(df['ecori_pos']==c_end) & (df['msei_pos']==c_start)]
        count_ls.append(tmp_df_bot.shape[0])

cut_sites['count'] = count_ls
cut_sites = cut_sites[cut_sites['count']>0]
cut_sites['response'] = 0.05
```
