Remove all fragments with internal cut sites.

```python
df = df[df['internal']==0]
```

## remove reads that map multiply from other genomes

```python
def get_hamming_distance(a, b):
    hamm = 0
    for i, j in zip(a, b):
        if i != j:
            hamm += 1
    return hamm/len(a)

threshold = 0.25 #rate of hamming distance
df_dupes = pd.DataFrame()

for i in range(0, df['length'].max()+1):
    tmp_df = df[df['length']==i].copy()
    seq_ls = tmp_df['seq'].to_list()
    gen_ls = tmp_df['genome'].to_list()
    
    hamm_ls = []
    for idxa, (seqa, gena) in enumerate(zip(seq_ls, gen_ls)):
        close_ones = 0
        for idxb, (seqb, genb) in enumerate(zip(seq_ls, gen_ls)):
            if idxa == idxb:
                pass
            elif gena == genb:
                pass
            else:
                hamm = get_hamming_distance(seqa, seqb)
                if hamm <= threshold:
                    close_ones += 1
        hamm_ls.append(close_ones)
    tmp_df.loc[:,'neighbors'] = hamm_ls
    df_dupes = pd.concat([df_dupes, tmp_df])
```

```python
df = df_dupes[df_dupes['neighbors'] == 0]
```

Calculate relative abundance by capturing the inter-taxa ratios within each fragment length. The average of these ratios will be used to determine the overall relative abundance of the taxa

```python
kmers_ls = df.columns.to_list()[17:-1]
```

```python
for kmer in kmers_ls:
    df[kmer+'_o'] = df[kmer] * df['observed']
    df[kmer+'_e'] = df[kmer] * df['counts']
```

```python
df
```

```python
kmers_ls = [i for i in kmers_ls if 'N' not in i]

diff_ls = []
kmers_dt = {}

for kmer in kmers_ls:
    o = df[kmer+'_o'].sum()
    e = df[kmer+'_e'].sum()
    diff_ls.append(o-e)
    kmers_dt[kmer] = o-e
```

```python
plt.figure(figsize=(20,5))
plt.xticks(rotation=90)
points=[idx for idx, i in enumerate(kmers_ls)]
plt.axhline(y=0, color = 'black')
for i in points:
    plt.axvline(x=i, color = 'whitesmoke', zorder=0)
plt.scatter(kmers_ls, diff_ls, s=100, c=diff_ls, zorder=1)
plt.cool()
plt.title('Change in kmer frequency in observed vs. expected fragment counts')
plt.xlabel('kmer')
plt.ylabel('observed - expected kmer count')
plt.show()
```
