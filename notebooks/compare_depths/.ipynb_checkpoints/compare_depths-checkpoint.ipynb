```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# motif only alignments (includes 100u/80c sims)

```python
df_pear = pd.read_csv('../../analysis/snipen_2021/depth_comparison/many_sims_pearson_corr.csv', index_col=0)
df_spear = pd.read_csv('../../analysis/snipen_2021/depth_comparison/many_sims_spearman_corr.csv', index_col=0)
```

```python
figure_names = ['real1', 'real2', 'real3',
                'A1', 'A2', 'A3',
                'B1', 'B2', 'B3',
                'C1', 'C2', 'C3',
                'D1', 'D2', 'D3',                  
                'E1', 'E2', 'E3']
figure_names_dt = {k:v for k, v in zip(df_pear.columns, figure_names)}
df_pear.columns = figure_names
df_spear.columns = figure_names
df_pear = df_pear.rename(index=figure_names_dt)
df_spear = df_spear.rename(index=figure_names_dt)
```

```python
sns.set(font_scale = 1.2)
cg = sns.clustermap(df_pear, cmap ="Spectral", linewidths = 0.10, annot=True, figsize=(16, 16));
plt.setp(cg.ax_heatmap.yaxis.get_majorticklabels(), rotation = 0)
cg
cg.savefig("many_pearson_corr_clustermap.png") 
cg.savefig("many_pearson_corr_clustermap.pdf") 
```

```python
sns.set(font_scale = 1.2)
sns.set(rc={'figure.figsize':(16,16)})
cg = sns.heatmap(df_pear, cmap ="Spectral", linewidths = 0.10, annot=True);
cg
plt.savefig("many_pearson_corr_heatmap.png") 
plt.savefig("many_pearson_corr_heatmap.pdf") 
```

```python
sns.set(font_scale = 1.2)
cg = sns.clustermap(df_spear, cmap ="Spectral", linewidths = 0.50, annot=True, figsize=(16, 16));
plt.setp(cg.ax_heatmap.yaxis.get_majorticklabels(), rotation = 0)
cg
cg.savefig("many_spearman_corr_clustermap.png") 
cg.savefig("many_spearman_corr_clustermap.pdf") 
```

```python
sns.set(font_scale = 1.2)
sns.set(rc={'figure.figsize':(16,16)})
cg = sns.heatmap(df_spear, cmap ="Spectral", linewidths = 0.10, annot=True);
cg
plt.savefig("many_spearman_corr_heatmap.png") 
plt.savefig("many_spearman_corr_heatmap.pdf") 
```

# motif only alignments

```python
df_pear = pd.read_csv('../../analysis/snipen_2021/depth_comparison/pearson_corr.csv', index_col=0)
df_spear = pd.read_csv('../../analysis/snipen_2021/depth_comparison/spearman_corr.csv', index_col=0)
```

```python
df_spear.drop(['null_0'], axis=1, inplace=True)
df_spear.drop(index='null_0', axis=0, inplace=True)
df_pear.drop(['null_0'], axis=1, inplace=True)
df_pear.drop(index='null_0', axis=0, inplace=True)
```

```python
sns.set(font_scale = 1.2)
cg = sns.clustermap(df_pear, cmap ="Spectral", linewidths = 0.0, annot=True, figsize=(16, 16));
plt.setp(cg.ax_heatmap.yaxis.get_majorticklabels(), rotation = 0)
cg
cg.savefig("pearson_corr_clustermap.png") 
cg.savefig("pearson_corr_clustermap.pdf") 
```
