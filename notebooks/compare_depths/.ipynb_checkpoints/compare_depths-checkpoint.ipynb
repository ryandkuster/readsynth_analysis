```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_pear = pd.read_csv('pearson_corr.csv', index_col=0)
df_pear
```

```python
df_spear = pd.read_csv('spearman_corr.csv', index_col=0)
df_spear
```

```python
f, ax = plt.subplots(figsize =(6, 6))
sns.heatmap(df_pear, ax = ax, cmap ="YlGnBu", linewidths = 0.0)
```

```python
sns.set(font_scale = 1.2)
cg = sns.clustermap(df_pear, cmap ="Spectral", linewidths = 0.0, annot=True, figsize=(16, 16));
plt.setp(cg.ax_heatmap.yaxis.get_majorticklabels(), rotation = 0)
cg
cg.savefig("pearson_corr_clustermap.png") 
cg.savefig("pearson_corr_clustermap.pdf") 
```
